```python
from main import main
import umap.umap_ as umap
```

```python
trained_model, full_dataset = main("kang")
```

```python
print(full_dataset.X.shape)
```

```python
z_basal = trained_model.encoder(full_dataset.X)
```

```python
reducer = umap.UMAP()
z_umap = reducer.fit_transform(z_basal)
```

```python
pert_idx = (full_dataset.obs["stim"].astype(str).str.upper() == "STIM").astype(int).values

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cov_idx = le.fit_transform(full_dataset.obs["cell_type"].astype(str))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(z_umap[:,0], z_umap[:,1], c=pert_idx, alpha=0.1)
plt.colorbar(label="Perturbation (0=CTRL, 1=STIM)")
plt.title("UMAP of z_basal colored by Perturbation")
plt.show()
```

```python
reducer = umap.UMAP()
z_ori = reducer.fit_transform(full_dataset.X)

plt.figure(figsize=(8, 6))
plt.scatter(z_ori[:,0], z_ori[:,1], c=pert_idx, alpha=0.1)
plt.colorbar(label="Perturbation (0=CTRL, 1=STIM)")
plt.title("UMAP of X colored by Perturbation")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(z_umap[:,0], z_umap[:,1], c=cov_idx, alpha=0.1)
plt.colorbar(label="Cell Type")
plt.title("UMAP of z_basal colored by cell type")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(z_ori[:,0], z_ori[:,1], c=cov_idx, alpha=0.1)
plt.colorbar(label="Cell Type")
plt.title("UMAP of X colored by cell type")
plt.show()
```
